```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np 
import sys

np.set_printoptions(threshold=sys.maxsize)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

df = pd.read_csv('company_address100.csv')
df.head()
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df.해외거래처주소)
document_distances = (tfidf_matrix * tfidf_matrix.T)
```

```python
tfidf_matrix.toarray()
```

```python
dd = document_distances.toarray()
dd
```

```python
for i in range(len(dd)):
    for j in range(len(dd)):
        if i != j and dd[i, j] > 0.5  and  dd[i, j] != 1.:
            print(i, j)
            print(df.iloc[i].해외거래처상호,df.iloc[i].해외거래처주소)
            print(df.iloc[j].해외거래처상호,df.iloc[j].해외거래처주소)
                        
```
