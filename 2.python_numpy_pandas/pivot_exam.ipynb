```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.pivot_table(index=['sex','day'], columns='smoker', values=['total_bill', 'tip'])
```

```python
tips.pivot_table(index=['sex'], columns='smoker', values=['tip'])
```

```python
tips.pivot_table(index=['sex'], columns='smoker', values=['tip']
                 , aggfunc=['sum','mean'], margins=True, fill_value=0)
```
